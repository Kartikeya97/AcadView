```python
import numpy as np
arr = np.random.random((10,1))
print(np.mean(arr))
```

```python
arr = np.random.random((20,1))
print(np.var(arr))
print(np.std(arr))
```

```python
A = np.random.random((10,20))
B = np.random.random((20,25))
C = np.matmul(A,B)
print(C)
print(np.sum(C))
```

```python
A = np.random.random((10,1))
def func(x):
    return 1/(1+np.exp(-x))
print(np.apply_along_axis(func,0,A))
```
